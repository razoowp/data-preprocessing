# data preprocessing- mean-median -handling missing values

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv('train.csv')
df.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
missing_value_percol = df.isnull().sum()/df.shape[0]*100
missing_value_percol
```

```python
missing_value_above20 = missing_value_percol[missing_value_percol > 20].index
missing_value_above20
```

```python
df2 = df.drop(columns=missing_value_above20)
df2.shape
```

```python
df3_just_numerical= df2.select_dtypes(include=['int64', 'float64'])
df3_just_numerical.head()
```

```python
df3_just_numerical.shape
```

```python
plt.figure(figsize=(25, 25))
sns.heatmap(df3_just_numerical.isnull())
```

```python
df3_just_numerical.isnull().sum()
```

```python
df3_just_numerical.isnull().sum().sum()
```

```python
df3_just_numerical[df3_just_numerical.isnull().any(axis=1)]
```

```python
missing_num_var = [var for var in df3_just_numerical.columns if df3_just_numerical[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(10, 10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_just_numerical[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
#NEW CODE
plt.figure(figsize=(10, 10))
sns.set()

for i, var in enumerate(missing_num_var):
    ax = plt.subplot(2, 2, i+1)
    sns.histplot(df3_just_numerical[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='red', ax=ax)
    sns.histplot(df3_just_numerical[var], bins=20, kde=False, color='blue', alpha=0.5, ax=ax)
    ax.set_xlabel(var)

plt.tight_layout()
plt.show()
```

# Mean


```python
df4_mean = df3_just_numerical.fillna(df3_just_numerical.mean())
df4_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()

for i, var in enumerate(missing_num_var):
    ax = plt.subplot(2, 2, i+1)
    sns.histplot(df3_just_numerical[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='red', ax=ax, label='Original')
    sns.histplot(df3_just_numerical[var], bins=20, kde=False, color='blue', alpha=0.5, ax=ax, label='Original Histogram')
    sns.histplot(df4_mean[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='green', ax=ax, label='Mean Imputation')
    ax.set_xlabel(var)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.set()

for i, var in enumerate(missing_num_var):
    ax = plt.subplot(2, 2, i+1)
    sns.histplot(df3_just_numerical[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='red', ax=ax, label='Original')
    #sns.histplot(df3_just_numerical[var], bins=20, kde=False, color='blue', alpha=0.5, ax=ax, label='Original Histogram')
    sns.histplot(df4_mean[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='green', ax=ax, label='Mean Imputation')
    ax.set_xlabel(var)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.set()

for i, var in enumerate(missing_num_var):
    ax = plt.subplot(2, 2, i+1)
    sns.kdeplot(df3_just_numerical[var], shade=True, color='red', ax=ax, label='Original')
    sns.kdeplot(df4_mean[var], shade=True, color='green', ax=ax, label='Mean Imputation')
    ax.set_xlabel(var)
    ax.legend()

plt.tight_layout()
plt.show()
```

# Median

```python
df5_median = df3_just_numerical.fillna(df3_just_numerical.median())
df5_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10, 10))
sns.set()

for i, var in enumerate(missing_num_var):
    ax = plt.subplot(2, 2, i+1)
    sns.histplot(df3_just_numerical[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='red', ax=ax, label='Original')
    #sns.histplot(df3_just_numerical[var], bins=20, kde=False, color='blue', alpha=0.5, ax=ax, label='Original Histogram')
    sns.histplot(df4_mean[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='green', ax=ax, label='Mean Imputation')
    sns.histplot(df5_median[var], bins=20, kde=True, line_kws={'linewidth': 5}, color='blue', ax=ax, label='Median Imputation')
    ax.set_xlabel(var)
    ax.legend()

plt.tight_layout()
plt.show()
```
